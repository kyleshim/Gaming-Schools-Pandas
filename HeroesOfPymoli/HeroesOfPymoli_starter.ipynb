```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df.tail()
#purchase_data_df['Age'].max()
```

## Player Count

* Display the total number of players


```python
players = len(purchase_data_df["SN"].drop_duplicates())
players
players_df = pd.DataFrame({"Total Players": [players]})
players_df.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
items = len(purchase_data_df["Item ID"].drop_duplicates())
total = float(purchase_data_df["Price"].sum())
purchases = len(purchase_data_df["Purchase ID"])
price = float(total/purchases)
purchase_analysis_df = pd.DataFrame({"Number of Unique Item": [items], 
                                     "Average Price": [price], 
                                     "Number of Purchases": [purchases], 
                                     "Total Revenue": [total]})
purchase_analysis_df["Average Price"] = purchase_analysis_df["Average Price"].map('${0:,.2f}'.format)
purchase_analysis_df["Total Revenue"] = purchase_analysis_df["Total Revenue"].map('${0:,.2f}'.format)
purchase_analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_demo = purchase_data_df[["SN","Gender"]]
gender_demo = gender_demo.rename(columns={"SN":"Total Count"})
gender_demo = gender_demo.groupby(['Gender']).nunique()
gender_demo['Percentage of Players']=gender_demo['Total Count']/players*100
gender_demo['Percentage of Players'] = gender_demo['Percentage of Players'].map('{0:,.2f}%'.format)
gender_demo_df = gender_demo[['Total Count', 'Percentage of Players']]
gender_demo_df = gender_demo_df.reindex(['Male', 'Female', 'Other / Non-Disclosed'])
gender_demo_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gen_purch_df = purchase_data_df[['Gender', 'Purchase ID', 'Price']]
gen_purch = gen_purch_df.groupby(['Gender'])
gen_purch_count = gen_purch.count()
gen_purch_avg = gen_purch.mean()
gen_purch_sum = gen_purch.sum()
gen_purch_count = gen_purch_count.drop(columns=['Purchase ID'])
gen_purch_count = gen_purch_count.rename(columns={'Price':'Purchase Count'})
gen_purch_uni = purchase_data_df[['Gender','SN']]
gen_purch_uni = gen_purch_uni.set_index(['Gender'])
gen_purch_uni = gen_purch_uni.groupby(['Gender']).nunique()
gen_purch_sum = gen_purch_sum.drop(columns=['Purchase ID'])
gen_purch_sum = gen_purch_sum.rename(columns={'Price':'Total Purchase Value'})
gen_purch_avg = gen_purch_avg.drop(columns=['Purchase ID'])
gen_purch_avg = gen_purch_avg.rename(columns={'Price':'Average Price'})
gen_purch_fin = pd.merge(gen_purch_count, gen_purch_sum, on='Gender')
gen_purch_fin = pd.merge(gen_purch_fin, gen_purch_avg, on='Gender')
gen_purch_fin['Avg Total Purchase Per Person'] = gen_purch_fin['Total Purchase Value']/gen_purch_uni['SN']
gen_purch_fin['Total Purchase Value'] = gen_purch_fin['Total Purchase Value'].map('${0:,.2f}'.format)
gen_purch_fin['Average Price'] = gen_purch_fin.loc[:,'Average Price'].map('${0:,.2f}'.format)
gen_purch_fin['Avg Total Purchase Per Person'] = gen_purch_fin['Avg Total Purchase Per Person'].map('${0:,.2f}'.format)
gen_purch_fin = gen_purch_fin.reindex(['Male', 'Female', 'Other / Non-Disclosed'])
gen_purch_fin = gen_purch_fin[['Purchase Count', 
                               'Average Price', 
                               'Total Purchase Value', 
                               'Avg Total Purchase Per Person']]
gen_purch_fin
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 50]
age_groups = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
age_df = purchase_data_df.loc[:,['Age', 'SN']]
age_df['Age Group'] = pd.cut(age_df['Age'], age_bins, labels=age_groups, include_lowest=True)
age_df = age_df.rename(columns={'SN':'Total Count'})
age_df = age_df[['Age Group', 'Total Count', 'Age']]
age_df = age_df.set_index(['Age Group'])
age_df = age_df.groupby('Age Group').nunique()
age_df['Percentage of Players']=age_df['Total Count']/players*100
age_df['Percentage of Players'] = age_df['Percentage of Players'].map('{0:,.2f}%'.format)
age_df = age_df.drop(columns=['Age'])
age_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_purch_df = purchase_data_df.loc[:,['Age', 'Purchase ID', 'Price', 'SN']]
age_bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 50]
age_groups = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
age_purch_df['Age Group'] = pd.cut(age_purch_df['Age'], age_bins, labels=age_groups, include_lowest=True)
age_purch = age_purch_df.groupby(['Age Group'])
age_purch_count = age_purch.count()
age_purch_count = age_purch_count.rename(columns={'Purchase ID':'Purchase Count'})
age_purch_avg = age_purch.mean()
age_purch_avg = age_purch_avg.rename(columns={'Price':'Average Purchase Price'})
age_purch_sum = age_purch.sum()
age_purch_sum = age_purch_sum.rename(columns={'Price':'Total Purchase Value'})
age_purch_uni = age_purch_df.groupby(['Age Group']).nunique()
age_purch_fin = pd.merge(age_purch_count, age_purch_sum, on='Age Group')
age_purch_fin = pd.merge(age_purch_fin, age_purch_avg, on='Age Group')
age_purch_fin = age_purch_fin[['Purchase Count', 'Average Purchase Price', 'Total Purchase Value']]
age_purch_fin['Avg Total Purchase Per Person'] = age_purch_fin['Total Purchase Value']/age_purch_uni['SN']
age_purch_fin['Total Purchase Value'] = age_purch_fin['Total Purchase Value'].map('${0:,.2f}'.format)
age_purch_fin['Average Purchase Price'] = age_purch_fin.loc[:,'Average Purchase Price'].map('${0:,.2f}'.format)
age_purch_fin['Avg Total Purchase Per Person'] = age_purch_fin['Avg Total Purchase Per Person'].map('${0:,.2f}'.format)

age_purch_fin
```
